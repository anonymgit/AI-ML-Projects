# Object Detection using OpenCV (Machine Learning HandsOn)

```python
import numpy as np  
import time
import cv2 as cv #image and video reading writing library
import imutils #resize
```

```python
cam=cv.VideoCapture(0)
time.sleep(1)
firstFrame=None
area=500
if not cam.isOpened():
    print("cannot open camera")
    exit()
while True:
    #captured frame by frame
    ret,frame=cam.read()
    
    #if frame is read correctly ret is True
    if not ret:
        print("can't receive frame (stream end?). Exiting...")
        break
    text="Normal/No movement Detected" #took this text for printing message while object detection
    frame = cv.resize(frame, (600, 600)) #resizing
    
    grayFrame=cv.cvtColor(frame,cv.COLOR_BGR2GRAY) #changed colored img to grayscale
    gaussianImg=cv.GaussianBlur(grayFrame,(21,21),0) #grayframe is frame img which we are getting through cam and here we are smoothing the img
    
    if firstFrame is None:
        firstFrame=gaussianImg #capturing 1st frame on 1st iteration
        continue
    imgDiff=cv.absdiff(firstFrame,gaussianImg) #absolute diff b/w is
    threshImg=cv.threshold(imgDiff,25,255,cv.THRESH_BINARY)[1]
    threshImg=cv.dilate(threshImg,None,iterations=2)
    
    cnts= cv.findContours(threshImg.copy(),cv.RETR_EXTERNAL,cv.CHAIN_APPROX_SIMPLE)
    cnts=imutils.grab_contours(cnts)
    for c in cnts:
        if cv.contourArea(c) < area:
            continue
        (x,y,w,h)=cv.boundingRect(c)
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        text="Moving Object detected"
    print(text)
    cv.putText(frame,text,(10,20), cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2)
    cv.imshow("camera Feed",frame)
    key=cv.waitKey(1) & 0xFF
    if key==ord("q"):
        break;
cam.release()
cv.destroyAllWindows()
```
